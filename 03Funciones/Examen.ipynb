```python
from typing import List

def gestionar_regalos() -> List[str]:
    """
    Solicita al usuario que introduzca regalos en una lista. Permite seguir agregando
    hasta que el usuario decida detenerse.
    """
    print("Bienvenido al programa de gestión de regalos")
    regalos: List[str] = []
    while True:
        regalo = input("Escribe un regalo: ")
        regalos.append(regalo)
        continuar = input("¿Quieres introducir más regalos? Escribe 'No' para terminar: ").strip().lower()
        if continuar == "no":
            break
    print(f"Lista de regalos: {regalos}")
    return regalos

def contar_regalo(regalos: List[str], regalo_buscado: str) -> int:
    """
    Cuenta cuántas veces aparece un regalo en la lista.
    """
    return regalos.count(regalo_buscado)

# Menú interactivo
def menu():
    print("=== Gestión de Regalos ===")
    regalos = gestionar_regalos()
    regalo_a_buscar = input("Introduce el regalo que deseas contar: ")
    print(f"El regalo '{regalo_a_buscar}' se pidió {contar_regalo(regalos, regalo_a_buscar)} veces.")

menu()
```

```python
def son_parecidas(cadena1: str, cadena2: str = "Feliz Navidad") -> bool:
    """
    Verifica si dos cadenas son parecidas:
    - Tienen la misma longitud.
    - Coinciden en las dos primeras y últimas letras, ignorando mayúsculas.
    """
    cadena1 = cadena1.lower()
    cadena2 = cadena2.lower()
    return (
        len(cadena1) == len(cadena2)
        and cadena1[:2] == cadena2[:2]
        and cadena1[-2:] == cadena2[-2:]
    )

def obtener_frase(texto: str, numero: int) -> str:
    """
    Devuelve la frase n-ésima de un texto con frases separadas por puntos.
    Si el índice está fuera de rango, devuelve 'Feliz Navidad'.
    """
    try:
        frases = texto.split(".")
        return frases[numero].strip()
    except IndexError:
        return "Feliz Navidad"

# Menú interactivo
def menu():
    print("=== Comparación de Cadenas ===")
    cadena1 = input("Introduce la primera cadena: ")
    opcion = input("¿Quieres comparar con otra cadena? (Sí/No): ").strip().lower()
    if opcion == "sí":
        cadena2 = input("Introduce la segunda cadena: ")
        print(f"¿Son parecidas? {son_parecidas(cadena1, cadena2)}")
    else:
        print(f"¿Es parecida a 'Feliz Navidad'? {son_parecidas(cadena1)}")

    print("\n=== Obtener Frase ===")
    texto_frases = "Feliz Año Nuevo.З Різдвом Христовим.Boas festas.Feliz Navidad."
    while True:
        try:
            n = int(input("Introduce el número de la frase que deseas obtener: "))
            print(obtener_frase(texto_frases, n))
            break  # Salir del bucle si la entrada es válida
        except ValueError:
            print("Error: Por favor, introduce un número entero válido.")

menu()
```

```python
from typing import List

def calcular_visitantes_totales(datos: List[List[int]]) -> List[int]:
    """
    Calcula el número total de visitantes por atracción.
    """
    return [sum(atraccion) for atraccion in datos]

def atraccion_mas_visitada(atracciones: List[str], totales: List[int]) -> str:
    """
    Determina cuál atracción tuvo más visitantes.
    """
    indice_max = totales.index(max(totales))
    return atracciones[indice_max]

def dia_y_atraccion_con_mas_visitantes(
    atracciones: List[str], datos: List[List[int]]
) -> str:
    """
    Encuentra el día y la atracción con más visitantes.
    """
    max_visitas = 0
    dia_max = 0
    atraccion_max = ""
    for i, atraccion in enumerate(datos):
        for j, visitantes in enumerate(atraccion):
            if visitantes > max_visitas:
                max_visitas = visitantes
                dia_max = j + 1
                atraccion_max = atracciones[i]
    return f"El mayor número de visitantes se produjo en {atraccion_max}, el día {dia_max}º, con {max_visitas} visitantes."

# Menú interactivo
def menu():
    atracciones = ["árbol", "bola", "muñeco", "noria"]
    visitantes = [
        [1000, 2000, 3000],
        [1500, 2500, 3200],
        [1200, 1800, 3900],
        [1100, 1900, 3100],
    ]
    print("\n--- Visitantes Totales ---")
    totales = calcular_visitantes_totales(visitantes)
    for atraccion, total in zip(atracciones, totales):
        print(f"Los visitantes totales de {atraccion} fueron {total}")
    
    print("\n--- Atracción Más Visitada ---")
    mas_visitada = atraccion_mas_visitada(atracciones, totales)
    print(f"La atracción más visitada fue {mas_visitada}.")
    
    print("\n--- Día con Más Visitantes ---")
    print(dia_y_atraccion_con_mas_visitantes(atracciones, visitantes))

menu()
```

```python
from typing import List

# Función para crear una lista unidimensional con enteros o cadenas
def crear_lista() -> List[str]:
    """
    Esta función permite crear una lista unidimensional a partir de los elementos que el usuario ingresa.
    Los elementos pueden ser cadenas de texto, separados por comas.
    
    Returns:
        List[str]: La lista de cadenas creada.
    """
    lista = input("Introduce los elementos de la lista separados por coma (ej. manzana,pera,uva): ")
    return lista.split(",")  # Se divide la cadena de texto por las comas para crear una lista


# Función para recorrer y mostrar los elementos de la lista
def recorrer_lista(lista: List[str]) -> None:
    """
    Esta función recorre e imprime los elementos de una lista unidimensional.
    
    Args:
        lista (List[str]): La lista que será recorrida.
    """
    print("\nRecorriendo la lista:")
    for i, elemento in enumerate(lista):
        print(f"Índice {i}: {elemento}")


# Función para agregar un nuevo elemento a la lista
def agregar_elemento(lista: List[str]) -> None:
    """
    Esta función agrega un nuevo elemento a la lista.
    
    Args:
        lista (List[str]): La lista a la que se le agregará un nuevo elemento.
    """
    elemento = input("Introduce el elemento a agregar: ")
    lista.append(elemento)
    print(f"Elemento '{elemento}' agregado a la lista.")


# Función para eliminar un elemento de la lista por su valor
def eliminar_elemento(lista: List[str]) -> None:
    """
    Esta función elimina un elemento de la lista basado en su valor.
    
    Args:
        lista (List[str]): La lista de la que se eliminará un elemento.
    """
    elemento = input("Introduce el elemento a eliminar: ")
    if elemento in lista:
        lista.remove(elemento)
        print(f"Elemento '{elemento}' eliminado de la lista.")
    else:
        print(f"Elemento '{elemento}' no encontrado en la lista.")


# Función para actualizar un elemento en una posición específica de la lista
def actualizar_elemento(lista: List[str]) -> None:
    """
    Esta función permite actualizar el valor de un elemento en un índice específico de la lista.
    
    Args:
        lista (List[str]): La lista en la que se actualizará un elemento.
    """
    indice = int(input("Introduce el índice del elemento a actualizar: "))
    if 0 <= indice < len(lista):
        nuevo_valor = input("Introduce el nuevo valor: ")
        lista[indice] = nuevo_valor
        print(f"Elemento en índice {indice} actualizado a '{nuevo_valor}'.")
    else:
        print("Índice fuera de rango.")


# Función para sustituir elementos de la lista por un valor específico
def sustituir_elementos(lista: List[str]) -> None:
    """
    Esta función sustituye todos los elementos de la lista que coincidan con un valor por otro valor.
    
    Args:
        lista (List[str]): La lista en la que se realizarán las sustituciones.
    """
    valor_a_reemplazar = input("Introduce el valor a sustituir: ")
    nuevo_valor = input("Introduce el nuevo valor: ")
    for i in range(len(lista)):
        if lista[i] == valor_a_reemplazar:
            lista[i] = nuevo_valor
    print(f"Elementos reemplazados por '{nuevo_valor}'.")


# Función para comparar cadenas dentro de la lista
def comparar_cadenas(lista: List[str]) -> None:
    """
    Esta función compara todas las cadenas dentro de la lista con una cadena proporcionada por el usuario.
    
    Args:
        lista (List[str]): La lista que contiene las cadenas a comparar.
    """
    cadena_a_comparar = input("Introduce la cadena con la que deseas comparar: ")
    for elemento in lista:
        if elemento == cadena_a_comparar:
            print(f"'{elemento}' es igual a '{cadena_a_comparar}'.")
        else:
            print(f"'{elemento}' NO es igual a '{cadena_a_comparar}'.")


# Función para crear una matriz bidimensional con enteros o cadenas
def crear_matriz() -> List[List[str]]:
    """
    Esta función permite crear una matriz bidimensional. El usuario ingresa cada fila como una lista separada por comas.
    
    Returns:
        List[List[str]]: La matriz bidimensional creada.
    """
    filas = int(input("¿Cuántas filas tendrá la matriz? "))
    columnas = int(input("¿Cuántas columnas tendrá la matriz? "))
    matriz = []
    for i in range(filas):
        fila = input(f"Introduce los elementos de la fila {i+1} separados por coma: ")
        matriz.append(fila.split(","))
    return matriz


# Función para recorrer y mostrar los elementos de una matriz bidimensional
def recorrer_matriz(matriz: List[List[str]]) -> None:
    """
    Esta función recorre e imprime los elementos de una matriz bidimensional.
    
    Args:
        matriz (List[List[str]]): La matriz que será recorrida.
    """
    print("\nRecorriendo la matriz:")
    for i, fila in enumerate(matriz):
        for j, elemento in enumerate(fila):
            print(f"Elemento en [{i}][{j}]: {elemento}")


# Función para agregar una nueva fila a la matriz
def agregar_fila(matriz: List[List[str]]) -> None:
    """
    Esta función agrega una nueva fila a la matriz.
    
    Args:
        matriz (List[List[str]]): La matriz a la que se le agregará una nueva fila.
    """
    fila = input("Introduce los elementos de la nueva fila separados por coma: ")
    matriz.append(fila.split(","))
    print(f"Fila {fila} agregada a la matriz.")


# Función para agregar una nueva columna a la matriz
def agregar_columna(matriz: List[List[str]]) -> None:
    """
    Esta función agrega una nueva columna a la matriz.
    
    Args:
        matriz (List[List[str]]): La matriz a la que se le agregará una nueva columna.
    """
    nueva_columna = input("Introduce los elementos de la nueva columna separados por coma: ").split(",")
    for i, fila in enumerate(matriz):
        fila.append(nueva_columna[i])
    print(f"Columna {nueva_columna} agregada a la matriz.")


# Función para eliminar una fila de la matriz
def eliminar_fila(matriz: List[List[str]]) -> None:
    """
    Esta función elimina una fila de la matriz según el índice proporcionado.
    
    Args:
        matriz (List[List[str]]): La matriz de la que se eliminará una fila.
    """
    fila_a_eliminar = int(input("Introduce el índice de la fila a eliminar: "))
    if 0 <= fila_a_eliminar < len(matriz):
        matriz.pop(fila_a_eliminar)
        print(f"Fila en índice {fila_a_eliminar} eliminada.")
    else:
        print("Índice de fila no válido.")


# Función para eliminar una columna de la matriz
def eliminar_columna(matriz: List[List[str]]) -> None:
    """
    Esta función elimina una columna de la matriz según el índice proporcionado.
    
    Args:
        matriz (List[List[str]]): La matriz de la que se eliminará una columna.
    """
    columna_a_eliminar = int(input("Introduce el índice de la columna a eliminar: "))
    if 0 <= columna_a_eliminar < len(matriz[0]):
        for fila in matriz:
            fila.pop(columna_a_eliminar)
        print(f"Columna en índice {columna_a_eliminar} eliminada.")
    else:
        print("Índice de columna no válido.")


# Función para interactuar con el usuario y mostrar un menú
def menu() -> None:
    """
    Muestra un menú interactivo para que el usuario realice operaciones con listas y matrices.
    """
    lista = []
    matriz = []

    while True:
        print("\n--- MENÚ PRINCIPAL ---")
        print("1. Crear una lista unidimensional")
        print("2. Crear una matriz bidimensional")
        print("3. Recorrer lista")
        print("4. Recorrer matriz")
        print("5. Agregar elemento a la lista")
        print("6. Agregar fila a la matriz")
        print("7. Agregar columna a la matriz")
        print("8. Eliminar elemento de la lista")
        print("9. Eliminar fila de la matriz")
        print("10. Eliminar columna de la matriz")
        print("11. Actualizar elemento de la lista")
        print("12. Sustituir elementos de la lista por valor")
        print("13. Comparar cadenas de texto en la lista")
        print("14. Salir")

        opcion = input("Selecciona una opción: ")

        if opcion == "1":
            lista = crear_lista()
        elif opcion == "2":
            matriz = crear_matriz()
        elif opcion == "3":
            if lista:
                recorrer_lista(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "4":
            if matriz:
                recorrer_matriz(matriz)
            else:
                print("Primero debes crear una matriz.")
        elif opcion == "5":
            if lista:
                agregar_elemento(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "6":
            if matriz:
                agregar_fila(matriz)
            else:
                print("Primero debes crear una matriz.")
        elif opcion == "7":
            if matriz:
                agregar_columna(matriz)
            else:
                print("Primero debes crear una matriz.")
        elif opcion == "8":
            if lista:
                eliminar_elemento(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "9":
            if matriz:
                eliminar_fila(matriz)
            else:
                print("Primero debes crear una matriz.")
        elif opcion == "10":
            if matriz:
                eliminar_columna(matriz)
            else:
                print("Primero debes crear una matriz.")
        elif opcion == "11":
            if lista:
                actualizar_elemento(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "12":
            if lista:
                sustituir_elementos(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "13":
            if lista:
                comparar_cadenas(lista)
            else:
                print("Primero debes crear una lista.")
        elif opcion == "14":
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida, intenta de nuevo.")


# Ejecutar el menú
menu()
```
